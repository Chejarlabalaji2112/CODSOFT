```python

roman = ["i", "ii", "iii", "iv"]
year =["2016","2017","2018","2019","2020","2021","2022", "2023"]
course = ["b.tech","mca","m.tech","mba"]
reg = ["r20","r19","r18","r16"]
month = "janfebmaraprmayjunjulaugsepoctnovdec"
exam_type = 'regularsupplementary'


import sqlite3
conn = sqlite3.connect("/content/res_links.db")
cur= conn.cursor()

col = []
```

```python

cur.execute("select * from sietk_res")
for each_text in cur.fetchall():
  pattern = []
  for each in each_text[0].split():
    if "for" in each and len("for") < len(each):
        index = each.index("i")
        pattern.append(each[index:])
    elif each in roman:
        pattern.append(each)
    elif each in course:
        pattern.append(each)
    elif each[1:-1] in reg:
        pattern.append(each[1:-1])
    elif each in exam_type:
        if "regu" in each:
            pattern.append(each)
            reg_index = len(pattern)
        elif "regular" in pattern:
            pattern[reg_index-1] = 'regular/supplementary'
        elif "supp" in each:
            pattern.append(each)

    elif each[:3] in month:
        temp = each.split("-")
        if temp[-1] in year:
            pattern.append(temp[0])
            pattern.append(temp[-1])
        else:
            pattern.append(each)
    elif each in year:
        pattern.append(each)
  pattern.append(each_text[1])
  col.append(pattern)
```

```python

count = 0
for i in col:
     count += 1
     print(count,i)
```
